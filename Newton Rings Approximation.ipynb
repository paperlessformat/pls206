```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Read data
df = pd.read_csv('Interference Rings 2.csv')
x = df['Newton Order'].values              # Fringe order (m)
D = df['Diameter_um'].values               # Measured diameters in µm

# Square the diameters for the linear relationship
y = D**2                                  # y = D^2 (µm²)

# Wavelength in µm (e.g., 0.5928 µm)
wavelength = 0.5928

# Linear regression: y = k*x + b
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Compute radius of curvature
R = slope / (4 * wavelength)

# Uncertainty in R (approximate)
R_err = std_err / (4 * wavelength)

print(f"Fitted slope: {slope:.4f} ± {std_err:.4f}")
print(f"Intercept: {intercept:.4f}")
print(f"Estimated radius of curvature (R): {R:.2f} ± {R_err:.2f} µm")
print(f"R² = {r_value**2:.5f}")

# Plot
plt.figure(figsize=(6,4))
plt.scatter(x, y, label='Data', color='blue')
plt.plot(x, slope*x + intercept, color='red',
         label=f'Fit: y = {slope:.1f}x + {intercept:.1f}')
plt.xlabel('Fringe Order (m)')
plt.ylabel(r'Diameter$^2$ (µm$^2$)')
plt.title('Newton Rings Fit for Radius of Curvature')
plt.legend()
plt.grid(True)
plt.show()
```
